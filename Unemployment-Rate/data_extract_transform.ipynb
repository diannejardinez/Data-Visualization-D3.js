```python
import pandas as pd
```

```python
# Load data
unemployment_df = pd.read_fwf('data/laucnty19.txt')
unemployment_df.head()
```

```python
#  Set header
unemployment_df.columns = unemployment_df.iloc[0]
unemployment_df= unemployment_df.iloc[1:]
unemployment_df.head()
```

```python
#  Rename Header
new_unemployment_df  = unemployment_df.rename(columns={'LAUS Code': 'LAUS-Code',
                                                   'Code': 'County-FIPS-Code',
                                                    'Code2': 'State-FIPS-Code',
                                                   'County Name/State Abbreviation': 'countyName',
                                                    'Year': 'year',   
                                                   'Force': 'laborForce',
                                                   'Employed': 'employed',
                                                   'Level Rate': 'Level-Rate'})
new_unemployment_df.head()
```

```python
# Split Level Rate to Unemployed Level and Unemployed Rate  
new_unemployment_df[['unemployedLevel','none1','none1','none1','none1','none1', 'rate']] = new_unemployment_df['Level-Rate'].str.split(' ',expand=True)
new_unemployment_df.head()
```

```python
#  Combine County and State FIPS code to id
new_unemployment_df["id"] = new_unemployment_df["County-FIPS-Code"] + new_unemployment_df["State-FIPS-Code"]
new_unemployment_df.head()
```

```python
#  Filter out data
final_unemployment_df = new_unemployment_df[['id', 'countyName', 'laborForce', 
                                             'employed', 'unemployedLevel', 
                                            'rate', 'year']]
final_unemployment_df.head(80)
```

```python
final_unemployment_df = final_unemployment_df.mask(final_unemployment_df.astype(object).eq('None')).dropna()
final_unemployment_df.head(79)
```

```python
#  Save to csv file
final_unemployment_df.to_csv('data/unemployment_2019.csv')
```
